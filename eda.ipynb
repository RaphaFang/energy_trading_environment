```python
con.execute("""
    SELECT
        timestamp_utc,
        strftime(timezone('Europe/Copenhagen', timestamp_utc), '%Y-%m-%d %H:%M %a') AS 丹麥當地,
        hour, dow, is_weekend, is_holiday, holiday_name,
        round(daylight_hours,1) AS 日照時數,
        round(sunrise_hour,2) AS 日出, round(sunset_hour,2) AS 日落,
        is_daylight AS 白天
    FROM calendar
    WHERE area='DK1'
    ORDER BY timestamp_utc
    LIMIT 48
    """).df()


df = con.execute("SELECT * FROM training WHERE area='DK1' ORDER BY timestamp_utc LIMIT 24").df()
print(len(df.columns), "欄:", list(df.columns))
df
```
